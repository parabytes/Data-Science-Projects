### Question 1.2

What are the chief problems with the following chart (what visualization principles does it violate)?

1. We should not use images because once we scale the image we end up scaling both height and width
2. The graph above uses Green as its primary color. We should avoid including colors like red and green in our charts, because 8% and 0.2% of women have some form of colorblindess. Furthermore, Red/Green colorblindness is most common. 
3. Attributes of color are being used to demonstrate intensity but as the percentage grow the color should get darker but here we can see the opposite.
4. Shape is not quantitatively perceived. Here the graph uses marijuana leaf and represents quantitative information.
5. Also, there is no scale.

### Question 1.3

Present an alternative and describe why it is better (what visualization principles does your version follow)?

```python
percentage = [36, 49, 44, 17]
width = 1/1.5
xs = list(range(len(percentage)))
x_labels = ["18-89 years", "30-49 years", "60-64 years", "65+ years"]

figure, axes = plt.subplots()

axes.set_title("Americans Who Have Tried Weed")
axes.bar(xs, percentage, width, color="dimgray", align="center")
axes.set_xticks(xs)
axes.set_xticklabels(x_labels)
axes.xaxis.grid(b=False)
axes.set_xlabel("Age Group")
axes.set_ylabel("Percentage")

plt.show()
plt.close()
```

```python
x_labels = ["18-89 years", "30-49 years", "60-64 years", "65+ years"]
percentage = [36, 49, 44, 17]

figure, axes = plt.subplots() 

width = 0.75

ind = np.arange(len(percentage))
axes.barh(ind, percentage, width, color="grey")
axes.set_yticks(ind+width/2)

axes.set_yticklabels(x_labels, minor=False)
axes.yaxis.grid(b=False)
axes.xaxis.grid(b=False)
plt.rcParams["figure.figsize"] = (12,10)
plt.title('Percentage of Americans Who Have Tried Weed')
plt.xlabel('Percentage')
plt.ylabel('Age Group') 

for i, v in enumerate(percentage):
    axes.text(v + 1, i, str(v), color='grey', fontweight="bold")
```

There are many ways to present the information. 
1. We can present it as a table, which is presented in 1.1
2. We can represent x axis as the age group and y axis as the percentage of Americans who have tried weed.
3. We can do the opposite of what we did in point 2.

In all of these cases we ensure that we are respectful of the colorblind and that the chart can be interpreted by them. Furthermore, we present information clearly. This information can be easily interpreted by anyone who takes a look at it. We do not use images here, we also have a scale. Also, bar charts are best used to represent categorical data, which is what we have. Another approach is below.

```python
figure = plt.figure(figsize=(7, 6))
x_labels = ["18-89 years", "30-49 years", "60-64 years", "65+ years"]
xs = [1, 2, 3, 4]
data = [36, 49, 44, 17]
axes = figure.add_subplot(1, 1, 1)
axes.plot(xs, data, "o", color="dimgray")
axes.vlines(xs, [0], data, linestyles='dotted', lw=2)
axes.set_xticks(xs)
axes.set_xlim((0, 5))
axes.set_ylim((0, 60))
axes.set_xticklabels(x_labels)
axes.xaxis.grid(False)
```

### Question 2.1

```python
# connect to the database
database = "contracts"
conn = sqlite3.connect(database)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
# fetch information from the database
vendors = pd.read_sql_query("SELECT * FROM contractors", conn)
actions = pd.read_sql_query("SELECT department, actions, dollars FROM actions", conn)
total_actions = actions.groupby("department")[["actions", "dollars"]].sum()
total_actions
```

### Question 2.3

Determine how many vendors received contracts from 1 department, 2 departments, etc. In other words, I want a table that looks like the following (numbers are absolutely made up):

| # Departments | Vendors |
|-------------|-----------|
| 1           | 1238      |
| 2           | 525       |
| 3           | 275       |
| ...         |           |
| n           | 2         |

Interpret the first two rows in words. What is the overall story? Now, using *percentages* (instead of these raw counts), make a chart that shows this story using the principles explored in class.

```python
departments = pd.read_sql_query("SELECT department, contractor_id FROM actions", conn)
contractor_id = pd.read_sql_query("SELECT contractor_id FROM actions", conn)
vendors = pd.read_sql_query("SELECT id, vendor FROM contractors", conn)

row_list = []
for index, rows in departments.iterrows():
    my_list = [rows.department, rows.contractor_id]
    row_list.append(my_list)
id_list = contractor_id["contractor_id"].unique().tolist()
id_dict = dict.fromkeys(id_list)

# how many contracts have each contractor received from each department
for item in row_list:
    if item[1] in id_dict:
        if id_dict[item[1]] is None:
            id_dict[item[1]] = 1
        else:
            id_dict[item[1]] += 1

x = pd.DataFrame(id_dict.items()).sort_values(by=1)
x_list = x[1].unique().tolist()
x_dict = dict.fromkeys(x_list)

# tally how many contractors received same number of contracts
for key in id_dict:
    if id_dict[key] in x_dict:
        if x_dict[id_dict[key]] is None:
            x_dict[id_dict[key]] = 1
        else:
            x_dict[id_dict[key]] += 1
data = pd.DataFrame(x_dict.items())
data.columns = ["# Departments", "Vendors"]
data
```

Above we have information about #Departments and Vendors. This data outlines how many vendors received contracts from number of departments. For instance, we see here that most vendors only received one contract from one department (could be any department), while only one vendor recieved contracts from 19 departments.

```python
df = pd.DataFrame(data,columns=['# Departments','Vendors'])
df.plot(x ='# Departments', y='Vendors', kind = 'bar', color="dimgray", width=1/1.5, figsize=(20, 10))
plt.show()
```

This chart helps with visualization, however our x-values are horizontal for some reason. Regardless it serves as a visualization device.

The main story we see here is how the U.S. Government spends or allocates its funds
